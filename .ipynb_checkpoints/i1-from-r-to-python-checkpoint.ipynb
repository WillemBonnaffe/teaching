# Main differences between R and Python

# 1. Indentation

In R, we mark the start and end of a code block with `{` and `}`, for example

```R
test <- T
if (test == T){
    print("Test is true")
} else {
    print("Test is not true")
}
```

In Python, those are omitted, to make the syntax clearer, and we use columns and indentation.

## Use tabulations for indentation

````python
test = False
if (test == True):
    print("Test is true")
else:
    print("Test is not true")
````

# 2. A lighter syntax, fewer parenthesis

`If`, `for`, and `while` have a different syntax.

```R
if (test == True){} --> if (test == True):
for (x in X){} --> for x in X:
while (condition){} --> while (condition):
```

# 3. Objects

The equivalent of R numeric and matrix objects in R are numpy arrays.
The list in R is the list in Python as well.

## Create a vector

````python
## Vector (one-dimensional object)
xx = np.array([1,2,3,4])
print(xx)
````

## Create a matrix

````python
## Matrix (two-dimensional object)
xxx = np.array([[1,2], [3,4]])
print(xxx)
````

## Create a list

````python
## List
x_list = [1,2,3,4]
print(x_list)
````

## Access elements

````python
## Vectors and matrices are indexed from 0 in python
print(x_list[0])
print(xx[0])
print(xxx[0,0])
````

## Check the type of objects

````python
## Check the type of objects
print(type(xx))
````

## Check the dimensions of objects

````python
## Check the dimensions of objects
print(len(x_list)) #
print(xx.shape)
print(xxx.shape)
````

# 4. Object basic operations

## Create vector from a list

````python
## Create vector from a list
x_list = [1,2,3,4]
xx = np.array(x_list)
print(xx)
````

## Create array from vector

````python
## Create array from vector
xxx = xx.reshape(2,2)
print(xxx)
````

## Reshape array

````python
## Reshape array
print(xxx.reshape(1,4)) # Row vector
print(xxx.reshape(4,1)) # Column vector
````

## Matrix multiplication

````python
u = np.array([[1,2],[3,4]])
v = np.array([[0.5],[2]])
print(np.matmul(u, v))
````

## Concatenate

```R
r <- c(u,v)
```

````python
u = np.array([1,2])
v = np.array([3,4])
r = np.concatenate((u,v), axis=0)
print(r)
````

## Bind vectors by rows

```R
r <- rbind(u,v)
```

````python
u = np.array([1,2])
v = np.array([3,4])
r = np.concatenate((u[None,:],v[None,:]), axis=0)
print(r)
````

## Bind vectors by columns

```R
r <- cbind(u,v)
```

````python
u = np.array([1,2])
v = np.array([3,4])
r = np.concatenate((u[:,None],v[:,None]), axis=0)
print(r)
````

## Bind matrices by rows

```R
r <- rbind(u,v)
```

````python
u = np.array([[1,2],[3,4]])
v = np.array([[5,6],[7,8]])
r = np.concatenate((u,v), axis=0)
print(r)
````

## Bind matrices by columns

```R
r <- cbind(u,v)
```

````python
u = np.array([[1,2],[3,4]])
v = np.array([[5,6],[7,8]])
r = np.concatenate((u,v), axis=1)
print(r)
````
